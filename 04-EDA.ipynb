```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
imdf = pd.read_csv('./imdf_Z.csv', index_col=0)
```

```python
imdf.head()
```

Let's check out the data types for the different elements without all the actors and directors being made into dummies once more.

```python
imdf.dtypes
```

Let's take a look at the relationships between box office earnings and critical scores.

```python
sns.jointplot(imdf['Box Office ($)']/1000000, imdf['Rotten Tomatoes Score'])
```

```python
sns.jointplot(imdf['Box Office ($)']/1000000, imdf['Metacritic Score'])
```

```python
sns.jointplot(imdf['Box Office ($)']/1000000, imdf['imdb Score'])
```

While we're here, lets see if there's any relationship between box office earnings and runtime.

```python
sns.jointplot(imdf['Box Office ($)']/1000000, imdf['Runtime'])
```

```python
sns.jointplot(imdf['Runtime'],imdf['Rotten Tomatoes Score'])
```

For good measure, let's check the correlation between the different critical metrics we are using.

```python
imdf['Rotten Tomatoes Score'].corr(imdf['Metacritic Score'])
```

```python
imdf['Rotten Tomatoes Score'].corr(imdf['imdb Score'])
```

```python
imdf['imdb Score'].corr(imdf['Metacritic Score'])
```
